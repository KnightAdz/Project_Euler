The sequence of triangle numbers is generated by adding the natural numbers. So the 7th triangle number would be 1 + 2 + 3 + 4 + 5 + 6 + 7 = 28. The first ten terms would be:

1, 3, 6, 10, 15, 21, 28, 36, 45, 55, ...

Let us list the factors of the first seven triangle numbers:

- 1: 1
- 3: 1,3
- 6: 1,2,3,6
- 10: 1,2,5,10
- 15: 1,3,5,15
- 21: 1,3,7,21
- 28: 1,2,4,7,14,28

We can see that 28 is the first triangle number to have over five divisors.

What is the value of the first triangle number to have over five hundred divisors?

```python
import numpy as np
```

```python
def triangular_number(n):
    # Recursive function to generate triangle numbers
    # will reach maximum recursion depth before we reach the numbers we need
    if n<=1:
        return 1
    else:
        return n+triangular_number(n-1)
```

```python
triangular_number(7)
```

```python
def num_divisors(n):
    # Return the number of divisors of number n
    
    # Start at 2 to include 1 and itself
    divs = 2
    for i in range(2,int(n**0.5)):
        if n%i==0:
            divs += 2
    if n%n**0.5==0:
        divs+=1
    return divs
```

```python
num_divisors(28)
```

```python
num_divisors(10)
```

```python
for n in range(1,100):
    print("{}\t{}\t{}".format(n,triangular_number(n),num_divisors(triangular_number(n))))
```

```python
n_divs = 0
n = 1
tri_num = 0
max_divs = 0
while n_divs < 500:
    tri_num += n
    n_divs = num_divisors(tri_num)
    n+=1
    if n_divs > max_divs:
        print("New max: {} divisors for {}".format(n_divs,tri_num))
        max_divs = n_divs
    
print(tri_num)
```
